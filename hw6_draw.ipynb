```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import pyplot as plt
import random
fig = plt.figure()
ax = Axes3D(fig)
x = np.linspace(0,10,100) # np.random.randint(20,size=10) 形如： array([4, 1, 4, 3, 8, 2, 8, 5, 8, 19])
y = np.linspace(0,10,100)

X, Y = np.meshgrid(x, y)
Z = X**4-2*(X**3)*Y+(X**2)*(Y**2)-5*X*(Y**3)-Y**4

plt.xlabel('x')
plt.ylabel('y')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='RdBu')
plt.savefig('C:\\Users\\Maggie Liu\\Desktop\\jyhw6.jpg',dpi=1000)
plt.show()
```

```python
d={}
values=[]
for i in range(200):
    X = random.uniform(-1.,1.)
#     Y = random.uniform(-1.,1.) #错误！feasible region需要那个等式成立啊！你这样是一个正方形而不是圆啊！
    Y = (1-X**2)**0.5
    Z = X**4-2*(X**3)*Y+(X**2)*(Y**2)-5*X*(Y**3)-Y**4
    d[(X,Y)]=Z
for v in d.values():
    values.append(v)
for i in range(199):
    if values[i+1]<values[i]:
        min_value = values[i+1]
def get_keys(d, value):
    return [k for k,v in d.items() if v == value]
print(min_value,get_keys(d,min_value))
```
